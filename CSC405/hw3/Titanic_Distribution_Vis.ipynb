```python
################################################# Q1 #######################################################################
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic=pd.read_excel("titanic.xls", "titanic")
# Group the data by passenger class (pclass) and count the number of survivors in each class
survived_by_pclass=titanic.groupby('pclass')['survived'].sum()
# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
survived_by_pclass.plot(kind='barh', color='red')
plt.title('Number of Passengers Survived by Pclass')
plt.xlabel('Number of Passengers')
plt.ylabel('Pclass')
plt.show()
```

We can see from the bar chart that pclass 1 exhibited the highest rate of survival.

```python
################################################# Q2 #######################################################################
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic=pd.read_excel("titanic.xls", "titanic")
# Create a pivot table to aggregate the number of survivors by class and gender
survived_by_class_gender=titanic.pivot_table(index='pclass', columns='sex', values='survived', aggfunc='sum')
# Plotting the vertical bar chart
plt.figure(figsize=(10, 6))
survived_by_class_gender.plot(kind='bar', color=['pink', 'skyblue'])
plt.title('Passengers Survived by Class and Gender')
plt.xlabel('Pclass')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.show()
```

This bar chart indicated that females had much higher rates of survival than males. Females in pclass 1 had a particularily higher rate of survival relative to other genders and pclasses.

```python
################################################# Q3 #######################################################################
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic=pd.read_excel("titanic.xls", "titanic")
# Create a pivot table to aggregate the number of survivors and deaths by class and gender
survived_dead_by_class_gender=titanic.pivot_table(index='pclass', columns='sex', values='survived', aggfunc=['sum', 'count'])
# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
survived_dead_by_class_gender.plot(kind='bar', stacked=True, color=['pink', 'skyblue'])
plt.title('Passengers Survived vs Dead by Class and Gender')
plt.xlabel('Pclass')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(['Survived - Female', 'Survived - Male', 'Dead - Female', 'Dead - Male'], title='Gender')
plt.show()
```

```python
################################################# Q4 #######################################################################
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic=pd.read_excel("titanic.xls", "titanic")
# Filter the dataset to include only passengers with cabins B30, C30, and D30
cabin_30_passengers=titanic[titanic['cabin'].str.contains('B30|C30|D30', na=False)]
# Plot histogram of fares paid by passengers in cabins B30, C30, and D30
plt.figure(figsize=(10, 6))
plt.hist(cabin_30_passengers['fare'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Fares Paid for Passengers in Cabins B30, C30, and D30')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

The histogram above suggests that the most common fare paid for a passenger in cabin 30 (B30, C30, or D30) was ~53

```python
################################################# Q5 #######################################################################
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic=pd.read_excel("titanic.xls", "titanic")
# Plot box plot of fares by passenger class
plt.figure(figsize=(10, 6))
titanic.boxplot(column='fare', by='pclass', grid=False)
plt.title('Box Plot of Fares by Passenger Class')
plt.suptitle('')  # Remove the automatic title from the boxplot
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks([1, 2, 3], ['1st Class', '2nd Class', '3rd Class'])
plt.show()
```
