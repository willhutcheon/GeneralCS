```python
#inner join
import pandas as pd

# Read the Excel file
titanic_df = pd.read_excel("titanic.xls")

print(titanic_df.columns)

# Convert to CSV
titanic_df.to_csv("titanic.csv", index=None, header=True)

# Read the CSV into a DataFrame
data = pd.read_csv("titanic.csv")

# Create a new DataFrame with a new column "passengerID" filled with row indices
data_with_passengerID = data.copy()
data_with_passengerID['PassengerID'] = data_with_passengerID.index + 1

# Create DataFrame containing columns PassengerId, Name, and Pclass
df1 = data_with_passengerID[['PassengerID', 'name', 'pclass']]

# Create DataFrame containing columns PassengerId, Survived, and Age
df2 = data_with_passengerID[['PassengerID', 'survived', 'age']]

# Perform an inner join on these two DataFrames using PassengerId as the key
merged_df = pd.merge(df1, df2, on='PassengerID', how='inner')

# Display the first 10 rows of the merged DataFrame
print(merged_df.head(10))
```

```python
#outer join
import pandas as pd

# Read the Excel file
titanic_df = pd.read_excel("titanic.xls")

print(titanic_df.columns)

# Convert to CSV
titanic_df.to_csv("titanic.csv", index=None, header=True)

# Read the CSV into a DataFrame
data = pd.read_csv("titanic.csv")

# Create a new DataFrame with a new column "passengerID" filled with row indices
data_with_passengerID = data.copy()
data_with_passengerID['PassengerID'] = data_with_passengerID.index + 1

# Create DataFrame containing columns PassengerId, Name, and Pclass
df1 = data_with_passengerID[['PassengerID', 'name', 'pclass']]

# Create DataFrame containing columns PassengerId, Survived, and Age
df2 = data_with_passengerID[['PassengerID', 'survived', 'age']]

# Perform an inner join on these two DataFrames using PassengerId as the key
merged_df = pd.merge(df1, df2, on='PassengerID', how='outer')

# Display the first 10 rows of the merged DataFrame
print(merged_df.head(10))
```

```python
#group by
import pandas as pd

# Read the Titanic data from the CSV file
data = pd.read_csv("titanic.csv")

# Task 1: Group by Pclass and calculate the average Fare and Age for each class
task1_result = data.groupby('pclass').agg({'fare': 'mean', 'age': 'mean'})
print("Task 1: Average Fare and Age for each Pclass")
print(task1_result)

# Task 2: Group by Embarked and Sex, and find the count of passengers for each combination
task2_result = data.groupby(['embarked', 'sex']).size().reset_index(name='Passenger Count')
print("\nTask 2: Passenger count for each combination of Embarked and Sex")
print(task2_result)
```

```python
#explore more
import pandas as pd

# Read the Titanic data from the CSV file
data = pd.read_csv("titanic.csv")

# Task 1: Calculate the proportion of passengers that survived by sex
task1_result = data.groupby('sex')['survived'].mean()
print("Task 1: Proportion of passengers that survived by sex")
print(task1_result)

# Task 2: Create age categories and calculate survival proportions by age category, pclass, and sex
# Define age categories
bins = [0, 14, 20, 64, 150]  # Age categories: children, adolescents, adult, senior
labels = ['Children', 'Adolescents', 'Adult', 'Senior']
data['Age_Category'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

# Calculate survival proportions by age category, pclass, and sex
task2_result = data.groupby(['Age_Category', 'pclass', 'sex'])['survived'].mean()
print("\nTask 2: Survival proportions by age category, pclass, and sex")
print(task2_result)
```
